```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Automobile_data.csv')
df
```

```python
droping_columns = ['symboling','normalized-losses','aspiration']
df = df.drop(columns=droping_columns)
df.head()
```

```python
df.info()
```

```python
# Replace '?' with NaN and convert columns to the appropriate data types
df['bore'] = df['bore'].replace('?', np.nan).astype('float64')
df['stroke'] = df['stroke'].replace('?', np.nan).astype('float64')
df['horsepower'] = df['horsepower'].replace('?', np.nan).astype('float64')
df['peak-rpm'] = df['peak-rpm'].replace('?', np.nan).astype('float64')
df['price'] = df['price'].replace('?', np.nan).astype('float64')
df.info()
```

```python
df.skew()
```

```python
# Fill missing values with mean or median
df['bore'].fillna(df.bore.mean(), inplace=True)
df['stroke'].fillna(df.stroke.median(), inplace=True)
df['horsepower'].fillna(df.horsepower.median(), inplace=True)
df['peak-rpm'].fillna(df['peak-rpm'].mean(), inplace=True)
df['price'].fillna(df.price.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
df1.head()
```

```python
df1.describe()
```

```python
num_df1 = df1.select_dtypes(include=['int','float'])
num_df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1_num_sc = pd.DataFrame(sc.fit_transform(num_df1), columns =  num_df1.columns)
df1_num_sc
```

```python
str_df1 = df1.select_dtypes(include=['object'])
str_df1
```

```python
str_df2 = str_df1[['fuel-type', 'num-of-doors', 'body-style', 'drive-wheels','engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system']]
str_df2
```

```python
dummy = pd.get_dummies(str_df2, drop_first=False)
dummy.head(3)
# take the numerical data as it is and take the catagrical data into numerical data
```

```python
df1_pre = pd.concat([df1_num_sc, dummy],axis=1)
df1_pre
```

```python
X = df1_pre.drop('price', axis=1)
X
```

```python
Y = df1.iloc[:,[-1]]
Y
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

```python
x_train.sort_index(ascending=True, inplace=True)
x_test.sort_index(ascending=True, inplace=True)
y_train.sort_index(ascending=True, inplace=True)
y_test.sort_index(ascending=True, inplace=True)
```

```python
import statsmodels.api as sm
```

```python
MLR_model1 = sm.OLS(y_train,x_train).fit()
MLR_model1.summary()
```

```python
y_test_pred = MLR_model1.predict(x_test)
y_test_pred.count()
```

```python
y_test_pred
```

```python
y_test_pred.sort_index(ascending=True, inplace=True)
```

```python
y_test.price
```

```python
y_test_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_true= y_test.price, y_pred = y_test_pred)
```
