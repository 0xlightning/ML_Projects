```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_excel('food.xlsx')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df=df.drop(columns=['ID','Delivery_person_ID'])
df
```

```python
df1=df.copy()
df1
```

```python
df1_num1=df1.select_dtypes(include=['float64','int64'])
df1_num1
```

```python
# df1_num1=df1_num.iloc[:,1:-1]
# df1_num1.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler() 
df1_num1_sc=pd.DataFrame(sc.fit_transform(df1_num1),columns=df1_num1.columns)
df1_num1_sc
```

```python
dummy=pd.get_dummies(df1.Type_of_order,drop_first=False)
dummy.head(4)
```

```python
dummy1=pd.get_dummies(df1.Type_of_vehicle,drop_first=False)
dummy.head(4)
```

```python
df1_pre=pd.concat([df1_num1_sc,dummy,dummy1],axis=1)
df1_pre.head(3)
```

```python
x=df1_pre
y=df1.iloc[:,[-1]]
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head(10)
```

```python
x_train.sort_index(ascending=True,inplace=True)
x_test.sort_index(ascending=True,inplace=True)
y_test.sort_index(ascending=True,inplace=True)
y_train.sort_index(ascending=True,inplace=True)
```

```python
import statsmodels.api as sm
MLR_model1=sm.OLS(y_train,x_train).fit() 
# OLS model gets trained
print(MLR_model1.summary())
```

```python
y_test_pred=MLR_model1.predict(x_test)
y_test_pred.count()
```

```python
y_test_pred
```

```python
y_test_pred.sort_index(ascending=True,inplace=True)
```
